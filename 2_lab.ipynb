# **Описание датасета**
Этот набор данных представляет собой список миллиардеров и компаний Forbes 2022 года. Датасет включает в себя место в рейтинге, имя и некоторую специальную информацию о человеке. В датасет включены 2667 записей.

# **Задание 1**
Вывод выбранного датасета. В нем присутствуют пропущенные данные и категориальные признаки: категория, страна, штат, город, пол. 

```python
import pandas as pd
import numpy as np

filename = '/content/drive/My Drive/Dataset/forbes_billionaires_2022.csv'
ds = pd.read_csv(filename)
pd.set_option('max_rows', 6)

pd.set_option('display.max_colwidth', None)
pd.set_option('display.float_format', '{:.2f}'.format)
ds.columns = ['Место', 'Имя', 'Возраст' , 'Состояние', 'Категория', 'Источник', 'Страна', 'Штат', 'Город', 'Организация', 'Selfmade', 'Пол', 'Дата рождения', 'Титул', 'Оценка благотворительности', 'Биография', 'О человеке']
ds = pd.DataFrame(ds)
display(ds)
```

# **Задание 2**
Для удаления пропусков в данных используется функция dropna. После удаления в датасете осталось 153 строки.

```python
ds = ds.dropna()
display(ds)
```

Кодирование категориальных признаков было сделано с помощью Label Encoder.
Данный тип кодирования является наиболее часто используемым, преобразование представляет собой однозначное соответствие число <-> уникальное значение категориального признака. 

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
ds['Категория'] = labelencoder.fit_transform(ds['Категория'].values)
display(ds)
```

Для просмотра списка закодированных значений используется функция labelencoder.classes_ . Реализация Label Encoder в sklearn прежде всего сортирует по алфавиту уникальные значения, потом присваивает им порядковый номер. Таким образом уникальный номер категориального признака -- это его индекс в данном массиве.

```python
labelencoder.classes_
```

Теперь закодируем все категориальные признаки.

```python
ds['Страна'] = labelencoder.fit_transform(ds['Страна'].values)
ds['Штат'] = labelencoder.fit_transform(ds['Штат'].values)
ds['Город'] = labelencoder.fit_transform(ds['Город'].values)
ds['Пол'] = labelencoder.fit_transform(ds['Пол'].values)
display(ds)
```

При нормализации меняем масштаб данных. Чаще всего масштабирование данных изменяется в диапазоне от 0 до 1. 

```python
from sklearn.datasets import load_iris
from sklearn import preprocessing
x_array = np.array(ds['Оценка благотворительности'])
normalized_arr = preprocessing.normalize([x_array])
ds['Оценка благотворительности'] = normalized_arr[0]


x_array = np.array(ds['Состояние'])
normalized_arr = preprocessing.normalize([x_array])
ds['Состояние'] = normalized_arr[0]
display(ds)
```
